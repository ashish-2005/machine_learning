# Feature Selection
* Feature Selection is important to dodge the <b>curse of dimensionality<b>
* It can be done in various ways such as dropping constant feature , pearson correlation ,

### Dropping Constant Feature
* As it sound , we just drop/remove all the constant or low variance feature with the help of <i><b>sklearn.feature_selection.VarianceThreshold</b></i>
* It return a list of binary each represents the column/feature indicate wether the column/feature have enough variance (we can set threshold)

```python
import pandas as pd
import numpy as np

#loading data
data = pd.read_csv('data.csv')
X = data.drop('TARGET',axis=1)
y = data['TARGET']

#identifying constant feature
from sklearn.feature_selection import VarianceThreshold
var_thr = VarianceThreshold(threshold=0) #any feature with 0 variance will be identified 
var_thr.fit(X)

constant_columns = [column for column in X.columns 
                    if column not in X.columns[var_thr.get_support()]]

X_new = X.drop(constant_columns,axis=1)

print(f'Features Before : {len(X.columns)}\nFeatures After : {len(X_new.columns)}\nFeature Dropped : {len(constant_columns)}')
#after applying above opp. we have selected 259 features out of 370 
```

### Correlation
* The idea is to replace multiple correlated feature with single feature .
* Example : if we have 2 feature ie. Income & Tax_paid these two are highly correlated feature so we can drop Tax_paid and now our model have one less feature to care about

```python
#creating sample dataset
X_sample = data.iloc[:,5:10] #taking only 10 features for visulization

import seaborn as sns
import matplotlib.pyplot as plt
corr_mat = X_sample.corr() # creating correlation matrix

plt.figure(figsize=(4,4))
sns.heatmap(corr_mat,annot=True,cmap=plt.cm.CMRmap_r)
plt.show()

#we can see there are indeed some feature who are correlated (see utl1 and utl3)
```

```python
#function to identify correlated feature 
def correlation(dataset,threshold): #threshold for minimum correlation
    col_corr = set() # set() to deal with duplicates
    corr_mat = dataset.corr()
    for i in range(len(corr_mat)):
        for j in range(i):
            if corr_mat.iloc[i,j] >= threshold:
                col_corr.add(corr_mat.columns[i])
    return col_corr
```

```python
corr_features = correlation(X,0.8)
X_new = X.drop(corr_features,axis=1) #dropping the correlated feature
print(f'Features Before : {len(X.columns)}\nFeatures After : {len(X_new.columns)}\nFeature Dropped : {len(corr_features)}')
```
