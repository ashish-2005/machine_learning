### ***Dimensionality Reduction Exercise***

##### 1)What are the main motivations for reducing a dataset’s dimensionality?What are the main drawbacks?
**ANS :-** The main motivation for dimensionality reduction is to tackle curse of dimensionality and get a more visualizable data, which states that things change drastically with increase in dimensions and ofc it make training extreamly slow.<br>
The major drawback has to be the **loss of information**.

##### 2)What is the curse of dimensionality?
**ANS:-** The curse of dimensionality is the fact that things change drastically for same thing in different dimensions which make it difficult of model to find a perfect solution/fit. for eg. the chance of a random point to lie close to border is 0.4% in 2d but it increaces to 99.9999% in 10,000 dimension , hence we can only imagin how a dataset with 1 million feature will vary

##### 3) Once a dataset’s dimensionality has been reduced, is it possible to reverse the operation? If so, how? If not, why?
**ANS:-** Yes, We can indeed reverse back the reduced dataset back to its original form but we will not get the exact dataset back because while dimensionality reduction we loose some varience/data which might effect the quality . while some algorithms (such as PCA) have a simple reverse transformation procedure that can reconstruct a dataset relatively similar to the original, other algorithms (such as T-SNE) do not

##### 4) Can PCA be used to reduce the dimensionality of a highly nonlinear dataset?
**ANS:-** No, PCA is a linear dimensionality reduction algorithm and work poorly on non-linear dataset ie. it looses too much information, however we can alway use the kernel trick along with PCA (kPCA) to use PCA over non-linear data.

##### 5) Suppose you perform PCA on a 1,000-dimensional dataset, setting the explained variance ratio to 95%. How many dimensions will the resulting dataset have?
**ANS:-** This is really hard to say as , it completely depends on the dataset if data is dependent on some feature a lot we might end up with just 5-10 or if data is extreamly scattred we even get 900-950 features

##### 6) In what cases would you use vanilla PCA, Incremental PCA, Randomized PCA,or Kernel PCA?
* Vanilla PCA is majorly used in the most basic/common conditions with no limits and exceptions 
* Incremental PCA is used when dataset is too huge that it can't be loaded on RAM at once so we use smaller batches of data , or if we are applying PCA in online learning , its slower than regular PCA
* Randomized PCA is also used for bigger dataset it reduces the time to find singular vectors through SVD by using random instances , hence faster than regular PCA
* kPCA is majorly used for highly nonlinear dataset or on those dataset in which PCA perform low or loose a lot of information

##### 7) How can you evaluate the performance of a dimensionality reduction algorithmon your dataset?
**ANS:-** A dimensionality reduction algorithm is said to work well if it eliminates a significant number of dimensions from the dataset without losing too much information.

##### 8) Does it make any sense to chain two different dimensionality reduction algorithms?
**ANS:-** It can absolutely make sense to chain two different dimensionality reduction algorithms. A common example is using PCA to quickly get rid of a large number of useless dimensions, then applying another much slower dimensionality reduction algorithm, such as LLE.

##### 9) Load the MNIST dataset and split it into a training set and a test set (take the first 60,000 instances for training, and the remaining 10,000 for testing). Train a Random Forest classifier on the dataset and time how long it takes, then evaluate the resulting model on the test set. Next, use PCA to reduce the dataset’s dimensionality, with an explained variance ratio of 95%. Train a new Random Forest classifier on the reduced dataset and see how long it takes. Was training much faster? Next evaluate the classifier on the test set: how does it compare to the previous classifier?

```python
from scipy.io import loadmat
mnist = loadmat("mnist-original.mat")
mnist_data = mnist["data"].T
mnist_label = mnist["label"][0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(mnist_data,mnist_label,
                                                 test_size=0.14285714285,random_state=42)
```

```python
from time import time
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

start_og = time()
rfc_og = RandomForestClassifier()#no hyperparameter as aim to find change in time and accuracy
rfc_og.fit(X_train,y_train)
end_og = time()
acc_og = accuracy_score(y_test,rfc_og.predict(X_test))

print(f'''
Accuracy   : {acc_og*100}%
Time taken : {int(end_og-start_og)} seconds
''')
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_train_red = pca.fit_transform(X_train)
X_test_red = pca.transform(X_test)

start_red = time()
rfc_red = RandomForestClassifier()#no hyperparameter as aim to find change in time and accuracy
rfc_red.fit(X_train_red,y_train)
end_red = time()
acc_red = accuracy_score(y_test,rfc_red.predict(X_test_red))

print(f'''
Accuracy   : {acc_red*100}%
Time taken : {int(end_red-start_red)} seconds
''')
```

the training time acctually increased insted of decreasing , this is because PCA work different for different datasets and algorithms .

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

start_og = time()
log_og = LogisticRegression(multi_class="multinomial", solver="lbfgs", random_state=42)
log_og.fit(X_train,y_train)
end_og = time()
acc_og = accuracy_score(y_test,log_og.predict(X_test))

print(f'''
Accuracy   : {acc_og*100}%
Time taken : {int(end_og-start_og)} seconds
''')
```

```python
start_red = time()
log_red = LogisticRegression(multi_class="multinomial", solver="lbfgs", random_state=42)
log_red.fit(X_train_red,y_train)
end_red = time()
acc_red = accuracy_score(y_test,log_red.predict(X_test_red))

print(f'''
Accuracy   : {acc_red*100}%
Time taken : {int(end_red-start_red)} seconds
''')
```

now we can see the power of PCA we traing 2x faster with a bit of loss in accuracy

```python
from sklearn.manifold import TSNE
from sklearn.pipeline import Pipeline

pca_plus_tsne = Pipeline([
    ('pca',PCA(n_components=0.95,random_state=42)),
    ('tsne',TSNE(n_components=2,random_state=42))
])

X_plotable = pca_plus_tsne.fit_transform(X_train)
```
